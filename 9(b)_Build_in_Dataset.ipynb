```python
# Import required library
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()

# Print type of dataset
print("type:")
print(type(iris))

# Print keys
print("\nkeys:")
print(iris.keys())

# Print type of data and target
print("\ntype of data and target:")
print(type(iris.data), type(iris.target))

# Print shape of data
print("\ndata shape:")
print(iris.data.shape)
```
